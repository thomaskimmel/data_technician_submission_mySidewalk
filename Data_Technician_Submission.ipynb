```python
#import necessary packages
import numpy as np
import pandas as pd
import datetime
from dateutil.relativedelta import relativedelta
from datetime import date
```

```python
#reformat datetimes
fire['Call Date'] = pd.to_datetime(fire['Call Date'])
fire['Received DtTm'] = pd.to_datetime(fire['Received DtTm'])
fire['On Scene DtTm'] = pd.to_datetime(fire['On Scene DtTm'])
```

```python
fire.head()
```

```python
#filter dates to only the past 12 months
start_date = '02-27-2019'
end_date = '02-27-2020'
```

```python
mask = (fire['Call Date'] > start_date) & (fire['Call Date'] <= end_date)
```

```python
fire = fire.loc[mask]
```

```python
len(fire)
```

```python
fire['Month'] = pd.to_datetime(fire['Call Date']).dt.to_period('M')
```

```python
#calculate the difference between emergency response on scene and call received in seconds 
fire['diff'] = fire['On Scene DtTm'] - fire['Received DtTm']
fire['diff'] = fire['diff']/np.timedelta64(1,'s')
```

```python
#group by month and battalion, calculate the top decile of response times 
grouped = fire.groupby(['Month', 'Battalion'], as_index=False)['diff'].quantile(.1)
```

```python
#rename columns
grouped_percentile = grouped.rename(columns={'diff': '90thPercentileResponseTimes', 'Battalion': 'EmergencyResponseDistrict'})
```

```python
#reorder columns to be consistent with example-table.csv
grouped_percentile = grouped_percentile[['EmergencyResponseDistrict', 'Month', '90thPercentileResponseTimes']]
grouped_percentile
```

```python
#export as csv
grouped_percentile.to_csv('SF_Fire_Response_Percentile.csv')
```
